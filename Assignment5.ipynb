# Q NO2: 2-D array named A

```python
A= maliha.random.randint(-7113,7113,size=((71,13)))    #71 rows and 13 columns
print(A)
```

```python
print(A.ndim)
```

```python
print(A[(A>0) &(A%2==0)])
```

```python
print(A[(A<0) &(A%2==0)])
```

```python
print(A[(A>0) &(A%2!=0)])
```

```python
print(A[(A<0) &(A%2!=0)])
```

```python
print(maliha.sum(A))
```

```python
print(maliha.sum(A,axis=1))
```

```python
print(maliha.sum(A,axis=0))
```

```python
print(A[(A%2==0)|(A%10==0)])
```

```python
print(A[(A%2!=0)&(A>0)&(A%11==0)])
```

```python
print(A[(A%2==0)].size)
```

```python
print(A[(A%2!=0)].size)
```

# QNO3:4-D array named A

```python
A= maliha.random.randint(-7113,7113,size=((((71,13,13,13)))))    #71 block and, 13 group,13 rows,13 column
print(A)
```

```python
print(A.ndim)
```

# POSITIVE EVEN

```python
print(A[(A>0)&(A%2==0)])          
```

# DIVISIBLE BY FIRST TWO DIGIT OF AG

```python
print(A[A%71==0])    # ag  7113 
```

# TOTAL NUM OF ZEROS

```python
print(A[A<0])
```
