## Q no 2: using for loop

## product has first char 'M'

```python
d={'Electronic':{'Laptop':{'Brand1':'Dell','Brand2':'HP','Brand3':'ASER','Brand4':'TOSHIBA'},
              'mobile':{'C1':'NOKIA','C2':'INFNIX','C3':'SAMSUNG','C4':'OPPO'}},
   'Makeup':{'Foundations':{'F1':'MISSROSE','F2':'HUDA BEAUTY','F3':'LOREAL','F4':'MAC'},
          'Lipstick':{'L1':'MEDORA','L2':'HUDA BEAUTY','L3':'LOREAL','L4':'FLARMAR'}}}
for k in d.values ():
    for v in k.values ():
        for i in v.values():
            if i[0]=='M':
                print(i)
        
```

## mobile product contained 'i'

```python
for k in d.values ():
    if 'mobile' in k:
        for v in k.values ():
            if v==k['mobile']:
                for i in v.values():
                    for ch in i:
                        if ch in 'I':
                            print(i)
```

## print  product cotained 'M' AND 'E'

```python
d={'Electronic':{'Laptop':{'Brand1':'Dell','Brand2':'HP','Brand3':'ASER','Brand4':'TOSHIBA'},
              'mobile':{'C1':'NOKIA','C2':'INFNIX','C3':'SAMSUNG','C4':'OPPO'}},
   'Makeup':{'Foundations':{'F1':'MISSROSE','F2':'HUDA BEAUTY','F3':'LOREAL','F4':'MAC'},
          'Lipstick':{'L1':'MEDORA','L2':'HUDA BEAUTY','L3':'LOREAL','L4':'FLARMAR'}}}
for k in d.values ():
    for v in k.values ():
        for i in v.values():
            if 'M' in i:
                print(i)
                for j in v.values():
                    if 'E' in j:
                        print(j)
```

## print foundations product have last character 'Y'


```python
d={'Electronic':{'Laptop':{'Brand1':'Dell','Brand2':'HP','Brand3':'ASER','Brand4':'TOSHIBA'},
              'mobile':{'C1':'NOKIA','C2':'INFNIX','C3':'SAMSUNG','C4':'OPPO'}},
   'Makeup':{'Foundations':{'F1':'MISSROSE','F2':'HUDA BEAUTY','F3':'LOREAL','F4':'MAC'},
          'Lipstick':{'L1':'MEDORA','L2':'HUDA BEAUTY','L3':'LOREAL','L4':'FLARMAR'}}}
for k in d.values ():
    if 'Foundations' in k:
        for v in k.values ():
            if v==k['Foundations']:
                for i in v.values():
                    if i[-1] in'Y':
                        print(i)
```

## print product less than 4

```python
d={'Electronic':{'Laptop':{'Brand1':'Dell','Brand2':'HP','Brand3':'ASER','Brand4':'TOSHIBA'},
              'mobile':{'C1':'NOKIA','C2':'INFNIX','C3':'SAMSUNG','C4':'OPPO'}},
   'Makeup':{'Foundations':{'F1':'MISSROSE','F2':'HUDA BEAUTY','F3':'LOREAL','F4':'MAC'},
          'Lipstick':{'L1':'MEDORA','L2':'HUDA BEAUTY','L3':'LOREAL','L4':'FLARMAR'}}}
for k in d.values():
    for v in k.values():
        for i in v.values():
            if(len(i)<4):
                print(i)
```
